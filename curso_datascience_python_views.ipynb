Importando o Matplotlib

```python
import matplotlib.pyplot as plt
```

Definição dos axies (eixos)



```python
x = [1, 2, 5]
y = [2, 3, 7]

#Construção do título
plt.title("Meu primeiro gráfico com Python")

#Legendas dos eixos
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

#Construção dos eixos utilizando especificação ".plot"
plt.plot(x, y)
plt.show()
```

Construção do gráfico de barras e nova estrutura de declarações

```python
#Declaração das variáveis
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]
titulo = "Gráfico de Barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Construção do título
plt.title(titulo)

#Legendas dos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)

#Construção dos eixos usando especificação ".bar"
plt.bar(x, y)
plt.show()
```

Construção de gráfico de barra com comparação entre eixos e legenda

```python
#Declaração das variáveis
x1 = [2, 4, 6, 8, 10]
y1 = [2, 3, 7, 1, 0]

x2 = [1, 3, 5, 7, 9]
y2 = [2, 4, 6, 8, 10]


titulo = "Gráfico de Barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Construção do título
plt.title(titulo)

#Legendas dos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)

#Construção dos eixos usando especificação ".bar" e exibindo legendas
plt.bar(x1, y1, label = "Grupo 1")
plt.bar(x2, y2, label = "Grupo 2")
plt.legend()
plt.show()
```

Scatterplot - Gráfico de dispersão + pontos ligados

```python
#Declaração das variáveis
x = [1, 2, 3, 4, 5]
y = [2, 3, 7, 1, 0]
titulo = "Scatterplot - Gráfico de Dispersão"
eixox = "Eixo X"
eixoy = "Eixo Y"

#Construção do título
plt.title(titulo)

#Legendas dos eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)

#Eixos usando especificação ".scatter" e ligando os pontos com ".plot"
plt.scatter(x, y, label = "Meus pontos", color='r', marker='H', s=100)
plt.plot(x, y, color='k', linestyle='-.')
plt.legend()
plt.show()
```
